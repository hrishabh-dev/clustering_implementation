```python
from sklearn.datasets import make_blobs

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.80, random_state=0)
```

```python
from sklearn.cluster import DBSCAN

# Instantiate DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the model to the data
dbscan.fit(X)

# Store the cluster labels
labels = dbscan.labels_
```

```python
from sklearn.cluster import KMeans

# Apply KMeans
kmeans = KMeans(n_clusters=4, random_state=0, n_init=10) # Assuming 4 clusters based on make_blobs
kmeans.fit(X)
kmeans_labels = kmeans.labels_
```

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
X_pca
```

```python
import matplotlib.pyplot as plt

# Create a figure with three subplots
plt.figure(figsize=(24, 6))

# DBSCAN plot
plt.subplot(1, 3, 1) # 1 row, 3 columns, 1st plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=50)
plt.title("DBSCAN Clustering Results")

# KMeans plot
plt.subplot(1, 3, 2) # 1 row, 3 columns, 2nd plot
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='o', s=50)
plt.title("KMeans Clustering Results")

# PCA plot
plt.subplot(1, 3, 3) # 1 row, 3 columns, 3rd plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], cmap='viridis', marker='o', s=50)
plt.title("PCA Results")


plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
```

```python
import plotly.express as px
import numpy as np

# Add a third dimension with a constant value for 3D visualization of 2D data
z_axis = np.zeros(X.shape[0])

# DBSCAN 3D plot
fig_dbscan_3d = px.scatter_3d(x=X[:, 0], y=X[:, 1], z=z_axis, color=labels, title="DBSCAN Clustering Results (3D Plotly)")
fig_dbscan_3d.show()

# KMeans 3D plot
fig_kmeans_3d = px.scatter_3d(x=X[:, 0], y=X[:, 1], z=z_axis, color=kmeans_labels, title="KMeans Clustering Results (3D Plotly)")
fig_kmeans_3d.show()
```
