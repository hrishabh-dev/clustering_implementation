```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')
```

```python
x,_=make_blobs(n_samples=500,centers=3,cluster_std=0.70,random_state=44)
plt.scatter(x[:,0],x[:,1],s=40)
plt.title("Unlabeled Synthetic Data")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
x
```

***K Means CLustering***

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=44)
label_kmeans=kmeans.fit_predict(x)
```

```python
label_kmeans
```

```python
plt.scatter(x[:,0],x[:,1],c=label_kmeans,cmap='rainbow',s=40)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='blue',marker='x',label='centroids') # position of centroids
plt.legend()
plt.title("K Means Clustering")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

**Elbow method**

```python
# wcss= within cluster sum of sqaures
```

```python
# it tries to find out the difference between the points and the clusters we will be having
```

Create the Elbow graph

```python
wcss=[]
k_range=range(1,11)
for i in k_range:
  kmeans=KMeans(n_clusters=i,random_state=44)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(k_range,wcss,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
